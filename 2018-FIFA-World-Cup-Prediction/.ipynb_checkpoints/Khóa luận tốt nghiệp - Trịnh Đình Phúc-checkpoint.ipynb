# Data Preparation (Chuẩn bị dữ liệu)

```python
data_ranking['rank_date'] = pd.to_datetime(data_ranking['rank_date']) 
data_ranking = data_ranking.replace({'IR Iran': 'Iran'})

data_histALL['date'] = pd.to_datetime(data_histALL['date']) # chuyển đổi kiểu dữ liệu time string thành timetamp
data_histALL =  data_histALL.replace({'Germany DR': 'Germany', 'China': 'China PR'})
```

```python
data_main = data_histALL.drop(labels=['city', 'country'], axis=1)

data_ranking = data_ranking[['rank_date', 'country_full', 'rank', 'rank_change']]
```

```python
# Form new column, 'previous_rank', and drop the column 'rank_change' for dataframe data_ranking
data_ranking['previous_rank'] = (data_ranking['rank' ]+ data_ranking['rank_change']).tolist() 
data_ranking = data_ranking.drop('rank_change', axis=1)

# Form new column, 'rank_change', for the dataframe data_main
data_main['home_win'] = data_main.home_score > data_main.away_score
```

### Gộp dữ liệu date và country (của cả đội sân nhà và sân khách)

```python
# Impute daily rankings
data_ranking = data_ranking.set_index(['rank_date']).groupby(['country_full'], group_keys=False).resample('D').first().fillna(method='ffill').reset_index()

# Merge data_ranking with data_main
data_main = data_main.merge(data_ranking, 
                        left_on=['date', 'home_team'], 
                        right_on=['rank_date', 'country_full'])

data_main = data_main.merge(data_ranking, 
                        left_on=['date', 'away_team'], 
                        right_on=['rank_date', 'country_full'], 
                        suffixes=('_home', '_away'))
```

```python
data_main.head(2)
```

```python
# Extract the useful features for modeling
data_main = data_main[['tournament', 'home_score', 'away_score','rank_home', 'previous_rank_home', 'rank_away', 'previous_rank_away', 'neutral', 'home_win']]
```

```python
data_main.head(2)
```

```python
# Create new column - 'score_diff'
data_main['score_diff'] = (abs(data_main.home_score - data_main.away_score)).tolist()
```

```python
#Xem độ tương quan giữa các features
plt.figure(figsize=(10,8))
corr = data_main.corr()
sns.heatmap(corr, xticklabels=corr.columns.values,cmap="Dark2",
            yticklabels=corr.columns.values, vmax=1,
            center=0,  annot= True, square=True)
```

# Normalizaion và PCA

```python
data_main.head(3)
```

```python
data_main[['home_score','away_score','rank_home','previous_rank_home','rank_away','previous_rank_away','neutral','score_diff']] = StandardScaler().fit_transform(data_main[['home_score','away_score','rank_home','previous_rank_home','rank_away','previous_rank_away','neutral','score_diff']])

sklearn_pca = sklearnPCA(n_components=8,svd_solver='full')
Y_sklearn = sklearn_pca.fit_transform(data_main[['home_score','away_score','rank_home','previous_rank_home','rank_away','previous_rank_away','neutral','score_diff']])
cum_sum = sklearn_pca.explained_variance_ratio_.cumsum()
sklearn_pca.explained_variance_ratio_[:10].sum()
cum_sum = cum_sum*100

fig, ax = plt.subplots(figsize=(10,6))
plt.bar(range(8), cum_sum, label='Cumulative _Sum_of_Explained _Varaince', color = 'red',alpha=0.5)
plt.xlabel('Số lượng K')
plt.ylabel('Score')
plt.title('Sử dụng phương pháp PCA để tránh multicollinearity')
```

## Dự đoán đội thắng 

```python
# Prepare training and test sets
x_train, x_test, y_train, y_test = train_test_split(data_main.drop(['home_win','home_score','away_score','score_diff'], axis=1), data_main.home_win.tolist(), test_size=0.3, random_state=42)
```

```python
x_train.head()
```

### Elbow analysis (phương pháp cùi trỏ)

```python
Nc = range(1, 20)
kmeans = [KNeighborsClassifier(i) for i in Nc]
kmeans
score = [kmeans[i].fit(x_train, y_train).score(x_train,y_train) for i in range(len(kmeans))] 
score #Opposite of the value of X on the K-means objective.
plt.figure(figsize=(10,7))
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
```

Dựa vào đồ thị ở trên ta có thể thấy đoạn từ 2.5 , 3 có độ dốc cao nhất. Tức là từ mức đó trở đi độ chính xác bị giảm rõ rệt. Nên ta sẽ dung k=3 cho thuật toán K-means

```python
classifiers = [
    KNeighborsClassifier(3),
    SVC(probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LogisticRegression()]

log_cols = ["Classifier", "Accuracy"]
log = pd.DataFrame(columns=log_cols)

acc_dict = {}

elapsed_time_each_algorithm = []
name_of_each_algorithm = []
for clf in classifiers:
    t = time.process_time()
    name = clf.__class__.__name__
    clf.fit(x_train, y_train)
    train_predictions = clf.predict(x_test)
    acc = accuracy_score(y_test, train_predictions)
    
    # lưu kết quả vào dictionary
    if name in acc_dict:
        acc_dict[name] += acc
    else:
        acc_dict[name] = acc
    elapsed_time = time.process_time() - t
    elapsed_time_each_algorithm.append(elapsed_time)
    name_of_each_algorithm.append(str(name))
    print("Thời gian chạy của thuật toán {0} là {1} giây \n".format(str(name),round(elapsed_time, 2)))
# Storing the results in a DataFrame to be visualized
for clf in acc_dict:
    acc_dict[clf] = acc_dict[clf]
    log_entry = pd.DataFrame([[clf, acc_dict[clf]]], columns=log_cols)
    log = log.append(log_entry)
```

```python
plt.figure(figsize=(12,6))
plt.xlabel('Độ chính xác (%)')
plt.title('So sánh độ chính xác giữa các model Machine Learning (%)')
sns.set_color_codes("muted")
sns.barplot(x='Accuracy', y='Classifier', data=log,color='cyan' ,saturation=.5)

plt.figure(figsize=(12,6))
plt.xlabel('Thời gian huấn luyện (s)')
plt.title('So sánh thời gian huấn luyện giữa các model Machine Learning (s)')
sns.set_color_codes("muted")
sns.barplot(x=elapsed_time_each_algorithm, y=name_of_each_algorithm,color='red',saturation=.5)

plt.show()
```

```python
log.Accuracy = log.Accuracy*100
log
```

```python
lr = LogisticRegression()
penalty_list = ['l1']
tol_list = [1e-3,1e-4,1e-5] # toleranre cho stopping_critetia
C_list = [0.5,1,5,10]
class_weight_list = ['balanced', None]
random_state_list = [12,42,122]
max_iter_list = [50,100,200]

param_grid_lr= dict(penalty=penalty_list,
                  tol=tol_list,
                  C=C_list, class_weight = class_weight_list,
                  random_state= random_state_list ,
                  max_iter = max_iter_list
                    )

grid_lr= GridSearchCV(lr, param_grid_lr,cv=5)
grid_lr.fit(x_train, y_train)
print(grid_lr.best_score_)
print(grid_lr.best_params_)
gre = pd.DataFrame(grid_lr.cv_results_)
(gre.loc[gre.mean_test_score!=0] # skip zero scored rows
     [['rank_test_score','mean_fit_time','mean_test_score'] # get these col
          +['param_{}'.format(i) for i in grid_lr.best_params_.keys()]]
    .sort_values('rank_test_score')) # sort by rank
gre.head(10)
```

```python
class_names = data_main.home_win.unique()
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Sử dụng Confusion Matrix")
    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
# tính confusion matrix
cnf_matrix = confusion_matrix(y_test, logR_model.predict(x_test))
np.set_printoptions(precision=2)
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Ma trận nhầm lẫn (Confusion matrix)')
plt.show()
print('Độ chính xác Accuracy: {}%\n'.format(int(np.diagonal(cnf_matrix).sum()/cnf_matrix.sum()*100)))
```

Dựa vào confusion matrix ta thấy ở lớp nhãn = 0 tức đội bị thua trên sân nhà, cứ n-trận diễn ra, thì có 70% trên tổng số n-trận đó đội sân nhà bị thua và ngược lại
